```python
import PIL as pt 
import numpy as np
```

```python
from google.colab import drive
drive.mount('/content/drive')
doc_img = pt.Image.open("/content/drive/MyDrive/notebook/larva.jpeg")
```

```python
#đọc ảnh ra màn hình 
doc_img
```

```python
#Chuyển bức ảnh từ cấu trúc dữ liệu của PIL sang NumPy array 
change_img = np.array(doc_img.getdata())

#Số chiều
print("số chiều", change_img.ndim)

#kích thước
print("kích thước", change_img.shape)
```

```python
#In nội dung của array thu được ở câu (4) ra màn hình.
change_img
```

```python
#chuyển ảnh màu sang ảnh xám 
new_img = doc_img.convert("L")
new_img
```

```python
#lưu vào máy
new_img.save("new_totoro.jpg")

#thu nhỏ ảnh
new_width  = 300
new_height = 300
small_img = doc_img.resize((new_width,new_height))
small_img
```

```python
#Chuyển ảnh màu sang Halftone 
new_img = doc_img.convert("1")
new_img
```

```python
#làm mờ ảnh 
from PIL import ImageFilter
new_img = doc_img.filter(ImageFilter.GaussianBlur(radius=20))
new_img
```

```python
#xoay ảnh 
n_img = doc_img.rotate((180))
n_img
```

```python
#pytorch 
import torch
```

```python
#tensor hạng 1 
ts_1 = torch.tensor([1,2,3,4,5])
print("Tensor hạng 1: ",ts_1)
print("hạng: ",ts_1.ndim)
```
