Câu 1: Sử dụng thư viện Numpy, hãy tạo một vector a có 7 chiều (tương ứng 7 phần tử) với giá
trị các phần tử được sinh bằng hàm (kết quả xuất ra màn hình):


*   np.arrange()
*   np.linspace()

*   np.logspace()

*   np.random.rand()

Ghi chú: sau mỗi lần sinh ra giá trị cho vector, hãy cho biết số chiều của vector được sinh ra và
số chiều của vector chuyển vị của vector đó; cho biết kích thước của vector a và vector chuyển vị
của vector a.
Gợi ý: tham khảo ndim và shape.










```python
import numpy as np

# np.arrange()

a =np.arange(1,8)
print(a,"\n")
print(a.ndim,"\n")
print(a.T.ndim,"\n")
print(a.shape,"\n")
print(a.T.shape)

# np.linspace()

b=np.linspace(0,6,7)
print(b,"\n")
print(b.ndim,"\n")
print(b.T.ndim,"\n")
print(b.shape,"\n")
print(b.T.shape,"\n")

# np.logspace()

c=np.logspace(1,7,7)
print(c)
print(c.ndim,"\n")
print(c.T.ndim,"\n")
print(c.shape,"\n")
print(c.T.shape,"\n")

# np.random.rand()

d=np.random.rand(7)
print(d)
print(d.ndim,"\n")
print(d.T.ndim,"\n")
print(d.shape,"\n")
print(d.T.shape)
```

Câu 2: Sử dụng thư viện Numpy, hãy tạo ra ma trận $A_{10x7}$ với giá trị các phần tử của ma trận A
được sinh bởi:


*   np.arrange()
*   np.linspace()

*   np.logspace()

*   np.random.rand()

Sau mỗi lần ma trận được tạo ra, hãy xuất kết quả ra màn hình và cho biết:



*   Số chiều và kích thước của ma trận A và ma trận $A^T$



```python
#ham arange()
a = np.arange(1,71).reshape(10,7)
print(a,"\n")
print(a.ndim,"\n")
print(a.T.ndim,"\n")
print(a.shape,"\n")
print(a.T.shape)

#ham linspace()
a = np.linspace(0,70,num=70).reshape(10,7)
print(a,"\n")
print(a.ndim,"\n")
print(a.T.ndim,"\n")
print(a.shape,"\n")
print(a.T.shape)

#hamlogspace()
a = np.logspace(0,70,num=70).reshape(10,7)
print(a,"\n")
print(a.ndim,"\n")
print(a.T.ndim,"\n")
print(a.shape,"\n")
print(a.T.shape)

#ham random.rand()
a = np.random.rand(10,7)
print(a)
print(a.ndim,"\n")
print(a.T.ndim,"\n")
print(a.shape,"\n")
print(a.T.shape)
```

Câu 4: Sử dụng thư viện Numpy, hãy:



*   Tạo ra 1 ma trận A10x15 với các giá trị được sinh ra bởi hàm logspace();

*   Hãy trích rút dòng thứ 3 của ma trận A (tương ứng với 1 vector hàng);
*   Hãy trích rút ra cột thứ 5 của ma trận A (tương ứng với 1 vector cột).

- Hãy tạo ra ma trận A15x10 với các giá trị sinh bởi hàm arrange()
- Áp dụng 5 hàm ở trên cho từng hàng của ma trận A
- Áp dụng 5 hàm ở trên cho từng cột của ma trận A

Gợi ý: tham khảo hàm apply_along_axis() đã học ở bài trước.

```python
#np.mean()
x = np.array([1,2,3,4,5])
print(np.mean(x))

#np.median()
X = np.array([2, 5, 3, 4,1, 7])
print(np.median(X))

#np.nanmedian()
y = np.array([2, 5, 3, 4,1, 7,np.nan])
print(np.nanmedian(y))

#np.var()
Y = np.array([1,2,4,6,7])
print(np.var(Y))

#np.std()
print(np.std(Y))

#GT trung binh
import numpy as np
a = np.arange(1,151).reshape(15,10)
print(a)
print(np.mean(a,axis=1))
print(np.mean(a,axis=0))

#GT trung vi
a = np.arange(1,151).reshape(15,10)
print(a)
print(np.median(a,axis=1))
print(np.median(a,axis =0))

#GT phuong sai
a = np.arange(1,151).reshape(15,10)
print(a)
print(np.var(a,axis=1))
print(np.var(a,axis =0))

#GT do lech chuan
a = np.arange(1,151).reshape(15,10)
print(a)
print(np.std(a,axis=1))
print(np.std(a,axis =0))
```
